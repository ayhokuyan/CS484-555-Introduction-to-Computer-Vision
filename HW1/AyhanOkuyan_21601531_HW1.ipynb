Useful links:

* https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_thresholding/py_thresholding.html






```python
# implement your function below.
def otsu_threshold(source_image):
    # This implementation is based on maximizing inter-class variance, rather 
    # than minimizing intra-class variance, which is computationally more efficient
    # ref: http://www.labbookpages.co.uk/software/imgProc/otsuThreshold.html
    hist = np.asarray(createHist(source_image))
    icvs = np.asarray([-1] * len(hist))
    num_pixel = len(source_image.ravel())
    gs_values = np.arange(256)
    for i in range(len(hist)):
        # no meaning in thresholsing from 0 or 255, there is no meaning in calculation
        if (i==0) or (i==len(hist)-1):
            pass
        # compute inter-class variances.
        else:
            w_b = np.sum(hist[:i-1])/num_pixel
            mean_b = np.sum(hist[:i-1]*gs_values[:i-1])/np.sum(hist[:i-1])
            w_f = np.sum(hist[i:])/num_pixel
            mean_f = np.sum(hist[i:]*gs_values[i:])/np.sum(hist[i:])
            # nan values are replaced with zeros 
            icvs[i] = w_b*w_f*(np.nan_to_num(mean_b)-np.nan_to_num(mean_f))**2
    best_thr = np.argmax(icvs)
    # binary thresholding
    binary_image = np.copy(source_image)
    binary_image[binary_image < best_thr] = 0
    binary_image[binary_image >= best_thr] = 1
    # return the binary image
    return binary_image


# See the OpenCV sample below that computes and compares three different thresholding technique on the same input image. 
# This code example uses OpenCV's built-in functions. HOWEVER, you need to implement OTSU's technique by yourself.
# the first thresholding below is the standard thresholding technique where we enter the threshold values manually,
# the second thresholding below is the OTSU's technique. 
# Change the image path according to your own Google drive structure!
img = cv.cvtColor(cv.imread(os.path.join(IMAGE_PATH, 'Q1.png')), cv.COLOR_BGR2GRAY)

# global thresholding
ret1,th1 = cv.threshold(img,10,255,cv.THRESH_BINARY)

# Otsu's thresholding
ret2,th2 = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)


# plot all the images and their histograms
plt.figure(figsize=(10,6))
images = [img, 0, th1,
          img, 0, th2]
titles = ['Original Image','Histogram','Manuel Thresholding',
          'Original Image','Histogram',"Otsu's Thresholding"]

plt.imshow(th1,'gray')

for i in [0,1]:
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
plt.show()

#Now lets show the result obtained with Otsu's technique in a larger frame. 
height, width = th2.shape[:2]
th2 = cv.resize(th2, (int(width/2), int(height/2))) 
cv2_imshow(th2) 
```

#### __Question 5 - Morphological Operators - Part A: Implement Dilation__ (15 points)

In  this section, you will first implement Dilation operator as a function. In your function, you will take two inputs (as matrices): (i) the input image as a matrix and (ii) a structuring element (also as a matrix), and produce a binary image (another matrix) as output. Note that the resulting image matrix should be of the same size as input image matrix. As before, in your function, your methods should be generic, that is it should work with grayscale image and structuring element of any size. Your method declaration should look as follows:

`dilated_image = dilation (source_image , structuring_el )`

* You should generate the structuring element (kernel) as a binary image with an arbitrary shape (you can use 3 by 3 square matrix with all 1s **for testing**). 

* Given the structuring element, your code should implement the dilation operation by using the definitions given in the course slides. Note that the structuring element should be created (as a matrix) outside and given as an input to the dilation/erosion codes so that your code can work with any kind of structuring element. You can assume that the origin is in the center, or get the coordinate of the origin as a separate input to your functions.

```python
## implement your python function for dilation  here
# implemetation is based on rectangular kernels with center being the origin.
#
def dilation(source_image, structuring_el):
    w_img,h_img = source_image.shape
    w_ker,h_ker = structuring_el.shape

    h_lim = int(np.floor(h_ker/2))
    w_lim = int(np.floor(w_ker/2))

    dilated_image = np.zeros(source_image.shape)

    #print(h_lim, w_lim)

    for x in range(w_img):
        for y in range(h_img):
            #choose appropriate valid kernel ranges for y axis
            if (y < h_lim):
                y_range_img = [0, y + h_lim]
                y_range_ker = [h_lim - y, h_ker]
            elif (y >= h_img - h_lim):
                y_range_img = [y - h_lim, h_img]
                y_range_ker = [0, h_lim + (h_img - y)]
            else:
                y_range_img = [y - h_lim, y + h_lim] 
                y_range_ker = [0, h_ker]
            #choose appropriate valid kernel ranges for x axis
            if (x < w_lim):
                x_range_img = [0, x + w_lim]
                x_range_ker = [w_lim - x, w_ker]
            elif (x >= w_img - w_lim):
                x_range_img = [x - w_lim, w_img]
                x_range_ker = [0, w_lim + (w_img - x) ]
            else:
                x_range_img = [x - w_lim, x + w_lim] 
                x_range_ker = [0, w_ker]
            rel_source = source_image[x_range_img[0]: x_range_img[1]+1, y_range_img[0]: y_range_img[1]+1]
            rel_str = structuring_el[x_range_ker[0]: x_range_ker[1], y_range_ker[0]: y_range_ker[1]]

            #print('src\n', rel_source.shape)
            #print('ker\n',rel_str.shape)
            sum = np.sum(rel_source*rel_str)
            if(sum > 0):
                dilated_image[x,y] = 1

    return dilated_image.astype(int)
```

```python
# implement your erosion operator here
# implemetation is based on rectangular kernels with center being the origin.
#
def erosion(source_image, structuring_el):
    w_img,h_img = source_image.shape
    w_ker,h_ker = structuring_el.shape

    h_lim = int(np.floor(h_ker/2))
    w_lim = int(np.floor(w_ker/2))

    eroded_image = np.zeros(source_image.shape)

    #print(h_lim, w_lim)

    for x in range(w_img):
        for y in range(h_img):
            #choose appropriate valid kernel ranges for y axis
            if (y < h_lim):
                y_range_img = [0, y + h_lim]
                y_range_ker = [h_lim - y, h_ker]
            elif (y >= h_img - h_lim):
                y_range_img = [y - h_lim, h_img]
                y_range_ker = [0, h_lim + (h_img - y)]
            else:
                y_range_img = [y - h_lim, y + h_lim] 
                y_range_ker = [0, h_ker]
            #choose appropriate valid kernel ranges for x axis
            if (x < w_lim):
                x_range_img = [0, x + w_lim]
                x_range_ker = [w_lim - x, w_ker]
            elif (x >= w_img - w_lim):
                x_range_img = [x - w_lim, w_img]
                x_range_ker = [0, w_lim + (w_img - x) ]
            else:
                x_range_img = [x - w_lim, x + w_lim] 
                x_range_ker = [0, w_ker]
            rel_source = source_image[x_range_img[0]: x_range_img[1]+1, y_range_img[0]: y_range_img[1]+1]
            rel_str = structuring_el[x_range_ker[0]: x_range_ker[1], y_range_ker[0]: y_range_ker[1]]

            #print('src\n', rel_source.shape)
            #print('ker\n',rel_str.shape)

            if np.all(rel_source == rel_str):
                eroded_image[x,y] = 1
    return eroded_image.astype(int)
```
